```python
dataset = dataset[['addr_state','annual_inc','issue_d','annual_inc_joint','application_type','dti','emp_length','emp_title','funded_amnt','funded_amnt_inv','grade','home_ownership','id','initial_list_status','loan_amnt','member_id','policy_code','pub_rec','purpose','sub_grade','term','title','total_acc','total_pymnt','verification_status_joint','zip_code','total_bal_il','il_util','verification_status','default_ind']]
```

```python
# Looking only those columns whose null count is greater than 0 percent, The below count of NAN is in percentage
null_col = dataset.isna().sum()
data_len = len(dataset)
null_col_info = null_col[null_col > 0].apply(lambda x: x/data_len)
```

```python
dataset.drop(null_col_info[null_col_info > 0.3].index, axis=1, inplace=True)
dataset.reset_index(inplace=True, drop=True)
```

```python
dataset.drop(dataset[dataset['emp_title'].isna()].index, inplace=True)
dataset.reset_index(drop=True, inplace=True)
```

```python
dataset.isna().sum()
```

```python
title_imputer = {}
for unq_purpose in dataset[dataset['title'].isna()]['purpose'].unique():
    impute = dataset[dataset['purpose'] == unq_purpose]['title'].value_counts().index[0]
    title_imputer[unq_purpose] = impute
    dataset['title'][dataset['purpose'] == unq_purpose] = dataset['title'][dataset['purpose'] == unq_purpose].fillna(impute)
```

```python
title_imputer
```

```python
#dataset['emp_title'] = dataset['emp_title'].apply(lambda x: " ".join(re.findall(r'[a-zA-Z]+\b',x)).upper()
```

```python
emp_length_imputer = {}

for user_emp_title in dataset[dataset['emp_length'].isna()]['emp_title'].value_counts().index:
    #print('----------------------------------------------\n',user_emp_title)
    avg_annual_inc = dataset[dataset['emp_title'] == user_emp_title]['annual_inc'].mean()
    
    user_purpose = dataset[dataset['emp_title'] == user_emp_title]['purpose'].value_counts().index[0]
    lower_annual_inc = int(avg_annual_inc) - 500
    upper_annual_inc = int(avg_annual_inc) + 500

    impute = dataset[(lower_annual_inc <= dataset['annual_inc']) & (dataset['annual_inc'] <= upper_annual_inc) & (dataset['purpose'] == user_purpose)]['emp_length'].value_counts().index[0]
    emp_length_imputer[user_emp_title] = impute
    
```

```python
emp_length_imputer
```

```python
for user_emp_title, imputer in emp_length_imputer.items():
    dataset['emp_length'][(dataset['emp_title'] == user_emp_title) & dataset['emp_length'].isna()] = dataset[(dataset['emp_title'] == user_emp_title)]['emp_length'].fillna(imputer)
```

```python
class_dist = dict(dataset['default_ind'].value_counts())
print(class_dist, class_dist[0] / class_dist[1], class_dist[0]/len(dataset), class_dist[1]/len(dataset))
```

```python
# Oversamplling Minor Class - 1
minor_class_data = dataset[dataset['default_ind'] == 1]
for _ in range(15):
    dataset = pd.concat([dataset, minor_class_data], ignore_index=True)
```

```python
class_dist = dict(dataset['default_ind'].value_counts())
print(class_dist, class_dist[0] / class_dist[1], class_dist[0]/len(dataset), class_dist[1]/len(dataset))
```

```python
# splitting train and test data
# Instruction : train (June 2007 - May 2015) and out-of-time test (June 2015 - Dec 2015) data
dataset['issue_d'] = pd.DatetimeIndex(dataset['issue_d'])
```

```python
train_data = dataset[('2007-06-01' <= dataset['issue_d']) & (dataset['issue_d'] <= '2015-05-01')]
train_data.to_csv('train_data.csv', index=False) # train data to csv
```

```python
test_data = dataset[('2015-06-01' <= dataset['issue_d']) & (dataset['issue_d'] <= '2015-12-01')]
test_data.to_csv('test_data.csv', index=False) # test data to csv
```

```python
trainclass_dist = dict(train_data['default_ind'].value_counts())
print(trainclass_dist, trainclass_dist[0] / trainclass_dist[1], trainclass_dist[0]/len(train_data), trainclass_dist[1]/len(train_data))
```

```python
testclass_dist = dict(test_data['default_ind'].value_counts())
print(testclass_dist, testclass_dist[0] / testclass_dist[1], testclass_dist[0]/len(test_data), testclass_dist[1]/len(test_data))
```
